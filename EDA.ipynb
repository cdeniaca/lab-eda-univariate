```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("amazon_uk_product_data.csv")

# Frequency table for product categories
category_freq = df['category'].value_counts()

# Top 5 most listed categories
top_5_categories = category_freq.head(5)
print("Top 5 most listed product categories:")
print(top_5_categories)

# Bar chart for product categories
plt.figure(figsize=(10, 6))
sns.barplot(x=category_freq.index[:10], y=category_freq.values[:10])
plt.title('Top 10 Product Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Pie chart for top categories
top_5_categories.plot.pie(autopct='%1.1f%%', figsize=(8, 8), legend=False)
plt.title('Proportions of Top 5 Product Categories')
plt.ylabel('')
plt.show()
```

```python
# Measures of centrality
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")

# Measures of dispersion
variance_price = df['price'].var()
std_dev_price = df['price'].std()
price_range = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Variance: {variance_price}")
print(f"Standard Deviation: {std_dev_price}")
print(f"Range: {price_range}")
print(f"IQR: {iqr_price}")

# Histogram for product prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Product Price Distribution')
plt.xlabel('Price ()')
plt.ylabel('Frequency')
plt.show()

# Box plot for price distribution
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['price'])
plt.title('Boxplot of Product Prices')
plt.xlabel('Price ()')
plt.show()
```

```python
from scipy.stats import kurtosis, skew

# Ensure the 'rating' column exists by printing the column names and checking for any leading/trailing spaces
df.columns = df.columns.str.strip()  # Clean up column names if needed

# Measures of centrality for ratings
mean_rating = df['rating'].mean()
median_rating = df['rating'].median()
mode_rating = df['rating'].mode()[0]

print(f"Mean Rating: {mean_rating}")
print(f"Median Rating: {median_rating}")
print(f"Mode Rating: {mode_rating}")

# Measures of dispersion for ratings
variance_rating = df['rating'].var()
std_dev_rating = df['rating'].std()
rating_range = df['rating'].max() - df['rating'].min()
iqr_rating = df['rating'].quantile(0.75) - df['rating'].quantile(0.25)

print(f"Variance: {variance_rating}")
print(f"Standard Deviation: {std_dev_rating}")
print(f"Range: {rating_range}")
print(f"IQR: {iqr_rating}")

# Skewness and Kurtosis
rating_skewness = skew(df['rating'])
rating_kurtosis = kurtosis(df['rating'])

print(f"Skewness: {rating_skewness}")
print(f"Kurtosis: {rating_kurtosis}")

# Histogram for product ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title('Product Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
